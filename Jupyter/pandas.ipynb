```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
df
```

```python
df.columns=['SL','SW','PL','PW']
```

```python
#Adding a new column
df['df_sl_sw']=df['SL']-df['SW']
df.tail()
```

```python
df.iloc[2:4,1:3]=np.nan
print(df.head())
dff=df.copy()
```

```python
df.describe()
```

```python
#Remove the row having nan values
df.dropna(inplace=True)
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df.describe()
```

```python
dff.describe()
```

```python
# we replaced the nan with the mean of the column
#if there is less  no of nan values then we can drop it otherwise will replace with mean
dff.SW.fillna(df.SW.mean(),inplace=True)
dff.PL.fillna(df.PL.mean(),inplace=True)
dff.head()
```

```python
# it's difficult to handle data with string values so we convert it to integer
dff['Gender']='Female'
dff.iloc[0:10,5]='Male'
print(dff)
def f(a):
    if a=='Male':
        return 0
    else:
        return 1
dff['sex']=dff.Gender.apply(f)

```
